```python
agency = []
contact_name = []
phone = []
alternate_cell = []

for i in range(1,40):
    
    
    website='https://www.zameen.com/Homes/Multan-15-' +str(i)+ '.html?area_min=209.03184000000002&area_max=1045.1592'
    
    # request to website
    response = requests.get(website)
   

    #soup varibale
    soup = BeautifulSoup(response.content, 'html.parser')
    
    
    
    
    
    
    
    #extracting infromation of agency, phone no. and contact name, these information is in call button inside script tags



    # 1- get script tag at 5th position that contains required information
    get_script_tag = soup.find_all('script')[5].get_text()


    # 2- finding tag with 'window.state = ' and using join to convert list into string
    if 'window.state = ' in get_script_tag:
        get_script_tag = ''.join([get_script_tag])
    
    
    #3 - parsing json
    get_script_tag = get_script_tag.split('window.state = ')[-1].split('}};')[0] + '}}'


    #4- load json
    get_json = json.loads(get_script_tag)


    #5- print json
    #print(json.dumps(get_json, indent=2))
    
    
    get_json = get_json['algolia']['content']['hits']
    
    
    #writing a json file 


    temp_file = json.dumps(get_json)
    with open ("json_data.json", "w") as file:
        file.write(temp_file)
        
        
    #reading json file


    with open('json_data.json', 'r') as json_file:
        json_load = json.load(json_file)

        #print(json_load)
    
    
    for c in range(1,25):
        try:
            agency.append(json_load[c]['agency']['name'])
        except:
            agency.append('')
            
        try:
            alternate_cell.append(json_load[c]['phoneNumber']['phone'])
        except:
            alternate_cell.append('')
            
            
        try:
            phone.append(json_load[c]['phoneNumber']['mobile'])
        except:
            phone.append('')
            
        try:
            contact_name.append(json_load[c]['contactName'])
        except:
            contact_name.append('')
```

```python
# Save data in python data_frame
df1 = pd.DataFrame({'Agency':agency,  'Contact_Name':contact_name, 'Phone':phone, 'Alternate_cell_no':alternate_cell})
df1
```

```python
title = []
loc = []
price = []
beds = []
baths = []
area = []


for i in range(1,40):
    
    
    website='https://www.zameen.com/Homes/Multan-15-' +str(i)+ '.html?area_min=209.03184000000002&area_max=1045.1592'
    
     # request to website
    response = requests.get(website)
   

    #soup varibale
    soup = BeautifulSoup(response.content, 'html.parser')
    
    results = soup.find_all('li', {'role' : 'article'})

    for result in results:
    
        # title
        try:
            title.append(result.find('h2', {'aria-label':'Title'}).get_text())
        except:
            title.append('')
    
        # location
        try:
            loc.append(result.find('div', {'aria-label':'Location'}).get_text())
        except:
            loc.append('')
        
        #area
        try:
            area.append(result.find('span', {'aria-label':'Area'}).get_text())
        except:
            area.append('')
    
        #price
        try:
            price.append(result.find('span', {'aria-label':'Price'}).get_text())
        except:
            price.append('')
        
        #beds
        try:
            beds.append(result.find('span', {'aria-label':'Beds'}).get_text())
        except:
            beds.append('')
    
        #baths
        try:
            baths.append(result.find('span', {'aria-label':'Baths'}).get_text())
        except:
            baths.append('')
        
 
 
```

```python
# Save data in python data_frame

df2 = pd.DataFrame({'Title':title, 'Location':loc, 'Area':area, 'Price':price, 'Beds':beds, 'Baths':baths})
df2
```

```python
# concatinate two data frames

df3 = pd.concat([df2,df1], ignore_index=True, axis=1)
df3.columns = [ 'Title', 'Location', 'Area', 'Price', 'Beds', 'Baths', 'Agency', 'Contact_Name', 'Phone', 'Alternate_cell_no',]
df3
```

```python
#replacing all empty cells with NaN
df3 = df3.replace('', np.nan)
```
